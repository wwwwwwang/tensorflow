```python
from tensorflow.examples.tutorials.mnist import input_data
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline

MNIST_data_folder="D:\\Users\\Madhouse\\tensorflow\\data\\mnist" 
mnist_images = input_data.read_data_sets(MNIST_data_folder, one_hot=False)

pixels,real_values = mnist_images.train.next_batch(10)

print("list for values loaded ", real_values)

example_to_visualize = 5

#print("element No. " + str(example_to_visualize + 1) + " of the list plotted:")

plt.figure(figsize=(12, 9)) 
for i in range(len(pixels)): 
    imagei = pixels[i] 
    showi = np.reshape(imagei, [28,28]) 
    plt.subplot(3,4,i+1) 
    plt.imshow(showi)

plt.show()
```
