```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import pandas as pd
%matplotlib inline

num_vectors = 1000
num_clusters = 4
num_steps = 500

x_values = []
y_values = []
vector_values = []

for i in range(num_vectors):
    if np.random.random() > 0.5:
        x_values.append(np.random.normal(0.4, 0.7))
        y_values.append(np.random.normal(0.2, 0.8))
    else:
        x_values.append(np.random.normal(0.6, 0.4))
        y_values.append(np.random.normal(0.8, 0.5))

vector_values = list(zip(x_values, y_values))
print("size = %d"%len(vector_values))

vectors = tf.constant(vector_values)
print(vectors.get_shape)

plt.plot(x_values, y_values, 'og', label='Input Data')
plt.legend()
plt.show()
```

```python
#init_op = tf.initialize_all_variables()
#sess = tf.Session()
#sess.run(init_op)

n_samples = tf.shape(vector_values)[0]
print(tf.shape(vector_values))
#print(sess.run(tf.shape(vector_values)))
print(n_samples)
#print(sess.run(n_samples))
random_indices = tf.random_shuffle(tf.range(0, n_samples))
print(random_indices)
#print(sess.run(random_indices))
print(tf.shape(random_indices))
#print(sess.run(tf.shape(random_indices)))

```

```python
def display_partition(x_values, y_values, assignment_values):
    labels = []
    colors = ["red", "blue", "green", "yellow"]
    for i in range(len(assignment_values)):
        labels.append(colors[assignment_values[i]])
    df = pd.DataFrame(dict(x=x_values, y=y_values, c=labels))
    fig, ax = plt.subplots()
    ax.scatter(df['x'], df['y'], c=df['c'])
    plt.show()


begin = [0,]
size = [num_clusters,]
size[0] = num_clusters
#print(begin)
#print(size)

centroid_indices = tf.slice(random_indices, begin, size)
centroids = tf.Variable(tf.gather(vector_values, centroid_indices))

expanded_vectors = tf.expand_dims(vectors, 0)
expanded_centroids = tf.expand_dims(centroids, 1)

vectors_subtration = tf.subtract(expanded_vectors, expanded_centroids)
euclidean_distances = tf.reduce_sum(tf.square(vectors_subtration), 2)
assignments = tf.to_int32(tf.argmin(euclidean_distances, 0))

partitions = tf.dynamic_partition(vectors, assignments, num_clusters)
    
update_centroids = tf.concat([tf.expand_dims(tf.reduce_mean(partition, 0), 0) for partition in partitions], 0)
#centroids = [tf.expand_dims(tf.reduce_mean(partition, 0), 0) for partition in partitions]
#update_centroids = tf.concat([centroids], 0)

change = tf.reduce_any(tf.not_equal(centroids, update_centroids))

with tf.control_dependencies([change]):
    update = tf.group(centroids.assign(update_centroids))

init_op = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init_op)

changed = True
iterNum = 0

while changed and iterNum < num_steps:
    iterNum += 1
    [changed, _] = sess.run([change, update])
    [centroid_values, assignment_values] = sess.run([centroids, assignments])
    
    print("iterNum=%d, changed=%r, centroid_values="%(iterNum, changed))
    print(centroid_values)
    display_partition(x_values, y_values, assignment_values)
```
