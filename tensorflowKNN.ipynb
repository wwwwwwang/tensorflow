```python
from tensorflow.examples.tutorials.mnist import input_data
import tensorflow as tf
import numpy as np
MNIST_data_folder="D:\\Users\\Madhouse\\tensorflow\\data\\mnist"
data_sets=input_data.read_data_sets(MNIST_data_folder,one_hot=True)

imageTrain, labelTrain = data_sets.train.next_batch(10000)
imageTest, labelTest = data_sets.test.next_batch(100)

iTrain = tf.placeholder(tf.float32, [None, 784])
iTest = tf.placeholder(tf.float32, [784])

def E_distance(x, y):
    #return tf.sqrt(tf.reduce_sum(tf.square(x-y), 2))
    return tf.sqrt(tf.reduce_sum(tf.square(x-y)))

#distance = tf.reduce_sum(tf.abs(tf.add(iTrain, tf.negative(iTest))), 1)
distance = tf.reduce_sum(tf.abs(iTrain-iTest), 1)
#distance = tf.sqrt(tf.reduce_sum(tf.square(tf.add(iTrain, tf.negative(iTest))), 1))
#distance = tf.sqrt(tf.reduce_sum(tf.square(iTrain-iTest), 1))
pred = tf.arg_min(distance, 0)

count = 0

init = tf.global_variables_initializer()

with tf.Session() as sess:
    sess.run(init)
    
    for i in range(len(imageTest)):
        pred_index = sess.run(pred, feed_dict={iTrain: imageTrain, iTest: imageTest[i]})
        print("In test No.", i, "prediction label: ", np.argmax(labelTrain[pred_index]),
              "true label: ", np.argmax(labelTest[i]))
        if np.argmax(labelTrain[pred_index]) == np.argmax(labelTest[i]):
            count +=1
    print("done!!")
    print("Accuary is %.2f%%"%(float(count)/len(imageTest)*100))
```
